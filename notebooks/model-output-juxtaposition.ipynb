```python
import pandas as pd
```

```python
output_with_transformer = pd.read_csv("eval-nocrf-with-transformer.txt", delimiter=" ", names=["token", "ground_truth", "with_transformer"])
output_no_transformer = pd.read_csv("eval-nocrf-no-transformer.txt", delimiter=" ", names=["token", "ground_truth", "no_transformer"])
```

```python
output = pd.concat([output_with_transformer, output_no_transformer["no_transformer"]], join="inner", axis=1)
list(output.iterrows())[0]
```

```python
with_transformer_correct = (output["ground_truth"] == output["with_transformer"])
no_transformer_correct = (output["ground_truth"] == output["no_transformer"])
```

```python
len(output)
```

```python
output[~with_transformer_correct & no_transformer_correct]
```

```python
output[with_transformer_correct & ~no_transformer_correct]
```

```python
output[~with_transformer_correct & ~no_transformer_correct]
```

```python
output.to_csv("eval-nocrf-merged.csv")
```
